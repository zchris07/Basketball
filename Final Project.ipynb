### Part 1: Finding and cleaning dataset  
The dataset I'm using is a huge basketball dataset with ~40 columns and ~20000 rows. It contains every NBA players' season stats for 60 seasons. Stats are per year.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
stats = pd.read_csv('stats')
stats = stats.drop(columns = ['Unnamed: 0', 'blanl', 'blank2'])
stats = stats[(stats['Year'] >= 1980) & (stats['G'] >= 20)]
pd.set_option('display.max_rows', 20, 'display.max_columns', 80)
stats
```

I included player's who played at least 20 games, and seasons starting from 1980.

### Part 2: Elementary Data Analysis 

```python
pd.set_option('display.max_rows', 40, 'display.max_columns', 80)
pd.DataFrame(stats.groupby('Year').mean()['3P%'])
```

```python
pd.DataFrame(stats.groupby('Year').mean()['3P'])
```

The change in the average players three-point percentage and and average three-pointers made during the entire season is shown in these two tables. The data shows the increasing importance of the three-pointer in the modern NBA.

```python
pd.DataFrame(stats.groupby('Year').mean()['3P']).diff().mean()
```

The average change in three-pointers made by any given player is +1.318 each season

### Part 3: Special Pandas

```python
pivot = stats.pivot_table(values=['PTS','3P','3P%','FT','Age','2P','DRB','ORB'],aggfunc=np.mean,index= 'Year')
pivot
```

```python
#average change season to season over 40 seasons
stats.pivot_table(values=['PTS','3P','3P%','FT','Age','2P','DRB','ORB'],aggfunc=np.mean,index= 'Year').diff().mean()
```

The pivot table provides more metrics for quantifying the evolution of basketball. We see that as players get better at shooting threes, regular two point shots/layups/dunks are used less. Similarly, free throws have been decreasing over time, and the average points scored per individual player have dropped. Offensive rebounds have also decreased significantly. Average age of the NBA player has remained relatively constant, at around 26-27 y/o.

### Part 4: Matplotlib

```python
plt.plot(pivot['3P'])
```

A plot like this could be used to reveal the direction in which the game is moving. Clearly the three-pointer is being used more and will probably continue to be used more in the future. Also lockout years can be easily seen in the graph.

```python
bird = stats[stats['Player'] == 'Larry Bird*']
bird_pivot = bird.pivot_table(values=['PTS','3P','3P%','FT','Age','2P','DRB','ORB'], aggfunc = np.mean, index= 'Year')
birdroll = pd.DataFrame(bird_pivot['3P'].rolling(3,center=True).mean())
plt.plot(birdroll)
```

A graph can also be used to track an individual players stats and highlight their areas of improvement.
